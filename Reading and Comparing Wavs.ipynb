```python
import soundfile as sf
import pandas as pd
import numpy as np
myColumns = ['ch1', 'ch2']
sigData = pd.DataFrame( columns=myColumns)

count = 1
chunk = 1024
#file = 'Dropouts/Frosty_ORIG24bitClean.wav'
file = 'Dropouts/Frosty.wav'
with sf.SoundFile(file, 'r') as f:
    size = 2*1024*1024 #f.frames #1024 #f.frames
    while f.tell() < size:
        data = f.read(chunk,dtype='int32')
        if count == 1: 
            sigData  = pd.DataFrame(data)
            count  += 1      
        else:
            sigData = pd.concat([sigData, pd.DataFrame(data)])
            count  += 1 
sigDataFrosty = sigData    
sigDataFrosty.reset_index(drop=True)
sigDataFrosty.columns = myColumns
print (sigDataFrosty.iloc[:,0].max(), sigDataFrosty.iloc[:,0].idxmax())
```
